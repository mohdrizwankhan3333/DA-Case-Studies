```python
plt.figure(figsize=(15,8))
plt.grid()
plt.title('Annual Income Dist.')
sns.histplot(df['Annual Income (k$)'], bins=20, kde=True, color='red')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(15,8))


plt.subplot(2,1,1)
sns.histplot(df['Age'], bins=20, kde=True, color='red')
plt.grid(True)
plt.title('Age Distribution')


plt.subplot(2,1,2)
sns.histplot(df['Spending Score (1-100)'], bins=20, kde=True, color='blue')
plt.grid(True)
plt.title('Spending Score Distribution')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
df.columns
```

```python
plt.subplot(2,1,1)
sns.kdeplot(data=df, x='Annual Income (k$)', hue='Gender', fill=True)
plt.title('Annual Income Distribution by Gender')
plt.subplot(2,1,2)

sns.kdeplot(data=df, x='Spending Score (1-100)', hue='Gender', fill=True)
plt.title('Spending Score (1-100) Distribution by Gender')
plt.tight_layout()
plt.show()
```
