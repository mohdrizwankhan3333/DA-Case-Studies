```python
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title('Customer Churn Distribution')
plt.xlabel('Churn Status')
plt.ylabel('Number of Customers')
plt.show()
```

```python
churn_counts = df['Churn'].value_counts()
plt.pie(churn_counts, labels=['Yes','No'], autopct='%1.1f%%')
plt.title('Churn Distribution')
plt.show()
```

```python
df.columns
```

```python
x=sns.countplot(data=df,x=df['gender'],hue='Churn')
plt.grid()
plt.title("Churn based on Gender")
plt.xticks(rotation=45)
plt.show()
```

```python
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Plot stacked bar chart
counts.plot(kind='bar', stacked=True)

plt.title('Churn by SeniorCitizen (Stacked Bar)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.grid(axis='x')
plt.legend(title='Churn')
plt.show()
```

```python
plt.title("Count of Senior Citizen Customer")
plt.grid()
sns.countplot(data=df,x='SeniorCitizen',palette='Set2')
```

```python
df['churn'] = df['Churn'].astype(str)
sns.histplot(
    data=df,
    x='tenure',
    hue='Churn',
    kde=True,
    palette='Set2',
    bins=30
)
plt.grid()
plt.title('Tenure Distribution')
plt.show()
```

```python
plt.title("Count of Customer by Contract")
plt.grid()
ax = sns.countplot(data=df, x='Contract', palette='Set2',hue='Churn')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sns.pairplot(data=df,palette='Set1')
```

```python
plt.title("Count of Customer by PaymentMethod")
plt.grid()
ax = sns.countplot(data=df, x='PaymentMethod', palette='Set2', hue='Churn')
ax.bar_label(ax.containers[0])
plt.xticks(rotation=45)
ax.legend(loc='upper right',fontsize='small')  # moves legend top right
plt.show()
```
