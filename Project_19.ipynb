```python
df2=df.groupby('state')['number'].sum().reset_index()
```

```python
df2
```

```python
plt.figure(figsize=(25,10))
ax=sns.barplot(data=df2,x='state',y='number',palette='Set3')
plt.grid()
for i in ax.containers:
  ax.bar_label(i)
plt.title('State Vs Number')
plt.show()
```
